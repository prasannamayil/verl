```python
import pandas as pd
import torch
import numpy as np
```

```python
base_dir = '/fast/pmayilvahanan/post_training/verl_checkpoints/active_grpo/'

freq = 2
results_dir = 'qwen2_1.5b_grpo_gsm8k_prioritization_freq_'+str(freq)+'_type_variance_alpha_2.0_epochs_4'
```

```python
# checks for advantage based prioritization with equal likely sampling for all kinds of groups
unique_indices = {}
for idx in range(1, 29, freq):
    try:
        index = torch.load(base_dir + results_dir + '/advantage_tracking/index_'+str(idx)+'.pt')
        unique_indices[idx] = np.sort(np.unique(index))
        print(idx, len(unique_indices[idx]))
    except:
        continue
    
```

```python
for idx in range(1, 29, freq):
    for idx2 in range(1, 29, freq):
        try:
            intersection = np.intersect1d(unique_indices[idx], unique_indices[idx2])
            print(idx, idx2, len(intersection))
        except:
            continue



```

```python
for idx in range(2, 29, freq):
    try:
        advantage = torch.load(base_dir + results_dir + '/advantage_tracking/advantages_train_step_'+str(idx)+'.pt')
        adv_nonzero = np.where(np.array(advantage['advantage_variances']) > 0.0)
        set_diff = set(unique_indices[idx+1]) - set(adv_nonzero[0])
        print(idx, len(set_diff))
    except:
        continue
```
